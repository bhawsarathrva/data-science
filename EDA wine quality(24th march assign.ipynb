Q2. How did you handle missing data in the wine quality data set during the feature engineering process? 
Discuss the advantages and disadvantages of different imputation techniques

Ans. To handle missing data in the wine quality data set during the feature engineering process, we use some methods which are disscussed below:
1. mean/median/mode imputation
Advantages: Simple and computationally efficient.
Disadvantages: Can distort the distribution of the data, especially if the missing data is not missing at random. Can introduce bias.
2. K-Nearest Neighbors (KNN) imputation
Advantages: Can capture more complex relationships between features, potentially leading to more accurate imputations.
Disadvantages: More computationally expensive than simpler methods. Sensitive to the choice of k.
3. Linear regression
Advantages: Can model linear relationships between features, leading to more accurate imputations.
Disadvantages: Assumes a linear relationship exists. Can be biased if the relationship is non-linear.
4. multiple imputation
Advantages: Accounts for uncertainty in the imputation process, leading to more realistic estimates of variability.
Disadvantages: More computationally expensive than other methods.
Deletion
5. Deletion
Advantages: Simple and avoids introducing bias.
Disadvantages: Can lead to a significant reduction in data, especially if there are many missing values. Can bias the results if missing values are not missing at random.

The best imputation technique depends on the specific dataset and the goals of the analysis. Consider the following factors:
Amount of missing data: If the amount of missing data is small, simple methods like mean/median imputation may be sufficient.
Distribution of the data: If the data is skewed or has outliers, more sophisticated methods like KNN or multiple imputation may be necessary.
Goals of the analysis: If you need to ensure that the analysis is not biased by the missing data, multiple imputation is often a good choice.

Q5. Load the wine quality data set and perform exploratory data analysis (EDA) to identify the distribution 
of each feature. Which feature(s) exhibit non-normality, and what transformations could be applied to 
these features to improve normality?
Ans. Transformtion to improve normality are:
1. Log Transformation 
2. Square Root Transformation

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats


wine_data = pd.read_csv("winequality-red.csv")  

# 1. Descriptive Statistics
print(wine_data.describe())

# 2. Histograms and Boxplots
for feature in wine_data.columns:
    plt.figure(figsize=(8, 4))
    plt.subplot(1, 2, 1)
    sns.histplot(wine_data[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    plt.subplot(1, 2, 2)
    sns.boxplot(y=wine_data[feature])
    plt.title(f'Boxplot of {feature}')
    plt.show()

# 3. Normality Test (Shapiro-Wilk Test)
for feature in wine_data.columns:
    stat, p = stats.shapiro(wine_data[feature])
    print(f'Shapiro-Wilk Test for {feature}:')
    print(f'  Statistic: {stat:.4f}')
    print(f'  P-value:   {p: 4f}')
    if p > 0.05:
        print(f'  The distribution of {feature} is likely normal.')
    else:
        print(f'  The distribution of {feature} is likely non-normal.')
    print()
```

```python
import numpy as np

# Apply log transformation to the 'alcohol' feature
wine_data['alcohol_log'] = np.log(wine_data['alcohol'] + 1)

# Plot the transformed feature
sns.histplot(wine_data['alcohol_log'], kde=True)
plt.title('Distribution of Alcohol (Log Transformed)')
plt.show()
```

Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of 
features. What is the minimum number of principal components required to explain 90% of the variance in 
the data?

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt


wine_data = pd.read_csv("winequality-red.csv")  

# 1. Separate features and target variable
X = wine_data.drop("quality", axis=1)  # Features
y = wine_data["quality"]  # Target variable

# 2. Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Perform PCA
pca = PCA()
pca.fit(X_scaled)

# 4. Calculate explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

# 5. Find the minimum number of components for 90% variance
cumulative_variance = np.cumsum(explained_variance_ratio)
min_components = np.argmax(cumulative_variance >= 0.90) + 1
print(f"Minimum number of components for 90% variance: {min_components}")

# 6. Plot explained variance ratio
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(explained_variance_ratio) + 1), cumulative_variance, marker='o')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Explained Variance by Number of Principal Components")
plt.grid(True)
plt.show()
```
