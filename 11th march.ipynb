Ans. T-Test:A t-test is a statistical test used to determine if there is a significant difference between the means of two independent groups. It assumes that the data is normally distributed and has equal variances.
Sample size: Suitable for small sample sizes (typically less than 30) when the population standard deviation is unknown.
Distribution: Uses the t-distribution, which has heavier tails than the normal (z) distribution, accounting for uncertainty introduced by estimating the population standard deviation from a small sample.
Example: You want to compare the average heights of two different groups of people. You collect data on the heights of 100 people in Group A and 100 people in Group B. You conduct a t-test to determine if there is a significant difference between the average heights of the two groups.

Z-Test: A z-test is a statistical test used to determine if there is a significant difference between a sample mean and a known population mean. It assumes that the data is normally distributed and has a known standard deviation.
Sample size: Suitable for larger sample sizes (typically greater than 30) or when the population standard deviation is known.
Distribution: ssumes a normal distribution, but the Central Limit Theorem allows for the approximation of normality with larger sample sizes.
Example: You want to determine if the average height of a population of people is 5 feet 9 inches. You collect data on the heights of 100 people from the population. You conduct a z-test to determine if there is a significant difference between the average height of the sample and the known population mean of 5 feet 9 inches.

Q2: Differentiate between one-tailed and two-tailed tests.
Ans. **One-Tailed Test:**

- In a one-tailed test, the researcher has a specific expectation about the direction of the relationship between the variables.
- It tests whether the data supports a specific hypothesis about the direction of the effect.
- For example, a researcher might hypothesize that a new educational program will lead to an improvement in students' test scores. The researcher would then conduct a one-tailed test to determine if the data supports this hypothesis.
- The critical value for a one-tailed test is located in only one tail of the sampling distribution.
- One-tailed tests are more powerful (have a higher chance of rejecting the null hypothesis when it is false) than two-tailed tests, but they are also more likely to make a Type I error (rejecting the null hypothesis when it is true).

**Two-Tailed Test:**

- In a two-tailed test, the researcher does not have a specific expectation about the direction of the relationship between the variables.
- It tests whether the data provides evidence of a statistically significant relationship between the variables, regardless of the direction of the effect.
- For example, a researcher might hypothesize that there is a relationship between smoking and lung cancer. The researcher would then conduct a two-tailed test to determine if the data supports this hypothesis.
- The critical value for a two-tailed test is located in both tails of the sampling distribution.
- Two-tailed tests are less powerful than one-tailed tests, but they are also less likely to make a Type I error.

Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for 
each type of error.
Ans. **Type 1 Error (False Positive)**

- A Type 1 error is rejecting the null hypothesis when it is actually true.
- In other words, the conclusion is that there is a statistically significant difference between the two variables when in reality there is no difference.

Example: 
A car manufacturer is testing a new model of car to see if it gets better gas mileage than its previous model. The null hypothesis is that the new model gets the same gas mileage as the previous model. The alternative hypothesis is that the new model gets better gas mileage than the previous model. The car manufacturer conducts a hypothesis test and finds that the new model does get better gas mileage than the previous model. However, the car manufacturer later discovers that the test results were flawed, and the new model actually gets the same gas mileage as the previous model. This is a Type 1 error.

**Type 2 Error (False Negative)**

- A Type 2 error is failing to reject the null hypothesis when it is actually false.
- In other words, concluding that there is no statistically significant difference between the two variables when in reality there is a difference.

Example: 
A pharmaceutical company is testing a new drug to see if it is effective in treating a particular disease. The null hypothesis is that the drug is not effective. The alternative hypothesis is that the drug is effective. The pharmaceutical company conducts a hypothesis test and finds that the drug is not effective. However, the pharmaceutical company later discovers that the test results were flawed, and the drug is actually effective. This is a Type 2 error.

Q4:  Explain Bayes's theorem with an example.
Ans. Bayes' theorem is a method of calculating conditional probabilities. It allows us to update our beliefs in light of new evidence.

To understand Bayes' theorem, it is helpful to first understand the concept of conditional probability. Conditional probability is the probability of an event occurring given that another event has already occurred. For example, the probability of it raining tomorrow given that there is a 30% chance of rain is 0.3.

Bayes' theorem states that the conditional probability of A given B is equal to the product of the probability of A and the probability of B given A, divided by the probability of B. Mathematically, this can be written as:

```
P(A|B) = (P(A) * P(B|A)) / P(B)
```

where:

* P(A|B) is the conditional probability of A given B
* P(A) is the probability of A
* P(B|A) is the probability of B given A
* P(B) is the probability of B
Example: 
Let's consider the following example. Suppose you are a doctor and a patient comes to you with a cough. You know that 10% of patients with a cough have pneumonia. You also know that 5% of patients without a cough have pneumonia.
What is the probability that the patient has pneumonia given that they have a cough?
To answer this question, we can use Bayes' theorem.
P(pneumonia|cough) = (P(pneumonia) * P(cough|pneumonia)) / P(cough)
We know that P(pneumonia) = 0.1 and P(cough|pneumonia) = 0.8. To find P(cough), we can use the law of total probability:
P(cough) = P(cough|pneumonia) * P(pneumonia) + P(cough|no pneumonia) * P(no pneumonia)
We know that P(cough|no pneumonia) = 0.05 and P(no pneumonia) = 0.9. Substituting these values into the equation, we get:
P(cough) = 0.8 * 0.1 + 0.05 * 0.9 = 0.086
Now we can substitute all of these values into Bayes' theorem:
P(pneumonia|cough) = (0.1 * 0.8) / 0.086 = 0.93
Therefore, the probability that the patient has pneumonia given that they have a cough is 0.93.

Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
Ans. **1. Confidence Interval**:
   - A confidence interval (CI) is a range of values estimated to contain the true value of a population parameter with a certain level of confidence.
   - This range is calculated from sample data and is used to make inferences about the population parameter.
   - The level of confidence is typically expressed as a percentage, such as 95% or 99%.

**2. Calculation of Confidence Interval**:
   - To calculate a confidence interval, we need:
     - **Sample Data**: A set of observations representing the population.
     - **Sample Mean ($\bar{x}$)**: The average value of the sample data.
     - **Sample Standard Deviation ($s$)**: A measure of the variability of the sample data.
     - **Confidence Level (Z-Score)**: A value obtained from the standard normal distribution that corresponds to the desired level of confidence.

**3. Formula for Confidence Interval**:
   - The general formula for the confidence interval is:
     - **Lower Boundary**: $\bar{x} - Z \times (s/\sqrt{n})$
     - **Upper Boundary**: $\bar{x} + Z \times (s/\sqrt{n})$

**4. Example**:
   - Suppose we have a sample of 100 grades from a statistics class, and the sample mean and standard deviation are 75 and 10, respectively.
   - To construct a 95% confidence interval for the mean grade in the population:
     - **Z-Score**: The z-score for a 95% confidence level is 1.96.
     - **Lower Boundary**: $75 - 1.96 \times (10/\sqrt{100}) = 72.04$
     - **Upper Boundary**: $75 + 1.96 \times (10/\sqrt{100}) = 77.96$
   - Therefore, we are 95% confident that the true mean grade in the population falls between 72.04 and 77.96.

It allows us to calculate the probability of an event occurring given prior knowledge of the event's probability and new evidence.

**Sample Problem:**

Consider a medical scenario where a patient is undergoing a test for a rare disease. The prior probability of the patient having the disease, denoted by P(Disease), is estimated to be 0.01 (1%).

Suppose there is a test available that is 95% accurate in detecting the disease when a patient has it, and 99% accurate in ruling out the disease when a patient does not have it. The test result comes back positive.

We want to calculate the probability that the patient actually has the disease given the positive test result, denoted by P(Disease | Test+).

**Solution:**

Step 1: Calculate the probability of a positive test result given that the patient has the disease:
P(Test+ | Disease) = 0.95 (95% accuracy)
Step 2: Calculate the probability of a positive test result given that the patient does not have the disease:
P(Test+ | No Disease) = 0.01 (1% false positive rate)
Step 3: Apply Bayes' Theorem:
P(Disease | Test+) = (P(Test+ | Disease) * P(Disease)) / P(Test+)
Step 4: Substitute the values from steps 1, 2, and 3:
P(Disease | Test+) = (0.95 * 0.01) / [0.95 * 0.01 + 0.01 * 0.99]
Step 5: Simplify and solve for P(Disease | Test+):
P(Disease | Test+) ≈ 0.095

where:

- CI is the confidence interval
- x̄ is the sample mean
- z is the z-score corresponding to the desired confidence level (for 95% confidence, z = 1.96)
- s is the sample standard deviation
- n is the sample size

In this case, we have:

- x̄ = 50
- s = 5
- n is not provided

- CI = 50 ± 1.96 * (5 / √100)
- CI = 50 ± 1.96 * (5 / 10)
- CI = 50 ± 1.96 * 0.5

Therefore, the 95% confidence interval for the sample data is (49.02, 50.98).

   - The margin of error in a confidence interval is the maximum amount by which the sample statistic (e.g., sample mean) may differ from the true population parameter (e.g., population mean) with a specified level of confidence. 
   - It represents the precision of the estimate, with a smaller margin of error indicating a more precise estimate. 

2. **Effect of Sample Size on Margin of Error:** 
   - The sample size plays a crucial role in determining the margin of error. Generally, as the sample size increases, the margin of error decreases. 
   - This is because a larger sample provides more data points, which leads to a more accurate representation of the population. 
   - Mathematically, the margin of error is inversely proportional to the square root of the sample size. 
   - Doubling the sample size reduces the margin of error by approximately 41%. 

3. **Example Scenario:** 
   - Suppose you want to estimate the average height of adults in the United States. You select a random sample of 100 adults and find that the average height is 5 feet 9 inches, with a 95% confidence interval of (5 feet 8 inches, 5 feet 10 inches). 
   - In this case, the margin of error is 1 inch (5 feet 10 inches - 5 feet 9 inches). 
   - If you had selected a larger sample, such as 400 adults, the margin of error would have decreased to approximately 0.5 inches, providing a more precise estimate of the average height.

Q12. A researcher is testing the effectiveness of two different teaching methods on student performance. 
Sample A has a mean score of 85 with a standard deviation of 6, while sample B has a mean score of 82 
with a standard deviation of 5. Conduct a hypothesis test to determine if the two teaching methods have a 
significant difference in student performance using a t-test with a significance level of 0.01.
Ans. 1. **State the Null and Alternative Hypotheses:**

Null Hypothesis (H0): µ1 = µ2 
Alternative Hypothesis (H1): µ1 ≠ µ2

where µ1 is the mean score of sample A and µ2 is the mean score of sample B.

2. **Calculate the Test Statistic:**

The test statistic for a two-sample t-test is given by:
t = (x̄1 - x̄2) / √(s1²/n1 + s2²/n2)

where x̄1 and x̄2 are the sample means, s1 and s2 are the sample standard deviations, and n1 and n2 are the sample sizes.

Plugging in the values, we get:

t = (85 - 82) / √(6²/5 + 5²/6)
t = 3 / √(36/5 + 25/6)
t = 3 / √(7.2 + 4.17)
t = 3 / 3.34
t = 0.9

3. **Determine the Degrees of Freedom:**

The degrees of freedom for a two-sample t-test is given by:
df = n1 + n2 - 2

Plugging in the values, we get:
df = 5 + 6 - 2
df = 9

4. **Find the P-value:**

The P-value is the probability of obtaining a test statistic as extreme as or more extreme than the observed test statistic, assuming the null hypothesis is true. We can use a t-distribution table or a statistical software to find the P-value.

Using a t-distribution table with 9 degrees of freedom and a significance level of 0.01, we find that the P-value is 0.393.

5. **Make a Decision:**

Since the P-value (0.393) is greater than the significance level (0.01), we fail to reject the null hypothesis. This means that we do not have sufficient evidence to conclude that there is a significant difference in student performance between the two teaching methods.

6. **Conclusion:**

Based on the results of the hypothesis test, we can conclude that there is no statistically significant difference in student performance between the two teaching methods at the 0.01 level of significance.
