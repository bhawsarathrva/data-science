Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans. Seaborn is a Python data visualisation library built on Matplotlib, another well-known visualisation package. Seaborn provides a high-level interface for making visually appealing and instructive statistics visuals. It's especially well-suited for analysing and visualising large datasets, making it popular among data analysts and scientists.

Seaborn library plots are:

1. Scatter plot: Scatter plots are used to visualize the relationship between two numerical variables. They help in identifying patterns, trends, clusters, or correlations in the data.

2. Line plot: Line plots are commonly used to display trends over time or other continuous intervals. They are useful for showing how a single variable changes in response to another variable.

3. Bar plot: Bar plots are used to compare values across different categories. They're effective in showing categorical data and can be used to display frequencies, counts, or aggregated values.

4. Histogram: Histograms are used to visualize the distribution of a single numerical variable. They show how data is spread across various bins or ranges and can help identify patterns like skewness, central tendency, and outliers.

5. Box plot: Box plots are used to display the distribution of data and identify potential outliers. They provide information about the median, quartiles, and potential extreme values in a dataset.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

fmri_data = sns.load_dataset('fmri')

sns.lineplot(data=fmri_data, x='timepoint', y='signal', hue='event', style='region')

plt.title("FMRI Signal Over Time by Event and Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

plt.legend(title="Event")

plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset("titanic")


plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title('Box Plot of Pclass vs Age')


plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Box Plot of Pclass vs Fare')

plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt


diamonds_data = sns.load_dataset("diamonds")

plt.figure(figsize=(10, 6))  
sns.histplot(data=diamonds_data, x="price", hue="cut", multiple="stack")

plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Frequency")

plt.legend(title="Cut")

plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset. 

```python
import seaborn as sns
iris= sns.load_dataset('iris')
sns.scatterplot(x = iris.sepal_length , y = iris.petal_length)
sns.relplot(x = iris.sepal_length , y = iris.petal_length , data = iris,hue = 'petal_length' )
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights_data = sns.load_dataset("flights")

flights_pivot = flights_data.pivot_table(index="month", columns="year", values="passengers")

plt.figure(figsize=(10, 6))
sns.heatmap(data=flights_pivot, annot=True, fmt="d", cmap="YlGnBu")

plt.title("Number of Passengers by Year and Month")
plt.xlabel("Year")
plt.ylabel("Month")

plt.show()
```
